```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array

# Load your trained model
model = tf.keras.models.load_model("baby_sleep_position_cnn_model.h5")
class_names = ['on_back', 'on_stomach', 'side']

# Define preprocessing function
def preprocess_frame(frame):
    img = cv2.resize(frame, (224, 224))
    img = img_to_array(img) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Flip and copy for display
    display_frame = cv2.flip(frame, 1)

    # Predict
    processed = preprocess_frame(display_frame)
    prediction = model.predict(processed)[0]
    label = class_names[np.argmax(prediction)]
    confidence = np.max(prediction) * 100

    # Display prediction on frame
    text = f"{label} ({confidence:.1f}%)"
    cv2.putText(display_frame, text, (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show frame
    cv2.imshow("Real-Time Sleep Position Classifier", display_frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
